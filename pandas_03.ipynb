# Binary Operation on Series

```python
import pandas as pd
#creating a series
df1=pd.Series([2,3,4,5,6], index=['a','b','c','e','t'])
df2=pd.Series([8,9,7,6,5], index=['c','d','f','g','k'])
print(df1,"\n",df2)
```

```python
c=df1.add(df2, fill_value=0)
print(c)
```

```python
d=df1.sub(df2,fill_value=0)
```

```python
print(d)
```
